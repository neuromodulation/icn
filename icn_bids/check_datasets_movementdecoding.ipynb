### BEIJING

```python
BIDS_PATH = r"C:\Users\ICN_admin\Documents\Decoding_Toolbox\Data\Beijing"
layout = bids.BIDSLayout(BIDS_PATH)
```

```python
layout.get_subjects()
```

```python
len(layout.get(extension='.vhdr', task='ButtonPress'))
```

```python
run_file = layout.get(extension='.vhdr', task='ButtonPress')[3]
entities = mne_bids.get_entities_from_fname(run_file.path)
path_vhdr = mne_bids.BIDSPath(subject=entities["subject"], session=entities["session"], task=entities["task"], \
    run=entities["run"], acquisition=entities["acquisition"], datatype="ieeg", root=BIDS_PATH)
raw_arr = mne_bids.read_raw_bids(path_vhdr)
```

```python
raw_arr.pick(["ecog"]).plot_psd(fmax=300, xscale="log")
plt.show()
```

```python
for run_file in layout.get(extension='.vhdr', task='ButtonPress'):
    entities = mne_bids.get_entities_from_fname(run_file.path)
    path_vhdr = mne_bids.BIDSPath(subject=entities["subject"], session=entities["session"], task=entities["task"], \
        run=entities["run"], acquisition=entities["acquisition"], datatype="ieeg", root=BIDS_PATH)
    raw_arr = mne_bids.read_raw_bids(path_vhdr)
    assert len(raw_arr.get_montage().dig) != 0
    assert raw_arr.get_data().shape is not None
```

```python
raw_arr.pick(["ecog", "seeg"]).plot(duration=2,scalings='auto')
plt.show()
```

```python
raw_arr.plot_psd(fmax=300, xscale="log")
plt.show()
```
