### leave one channel out CV

```python
def leave_one_ch_out(loc="ECOG", lat_="r2_con"):
    r2 = np.array(df_all[(df_all["loc"] == loc)][lat_])
    dist = np.array(df_all[(df_all["loc"] == loc)]["dist_con"])
    pred_arr = np.zeros([r2.shape[0]])
    true_arr = np.zeros(r2.shape[0])
    for i in range(r2.shape[0]):
        r2 = list(df_all[(df_all["loc"] == loc)][lat_])
        dist = list(df_all[(df_all["loc"] == loc)]["dist_con"])
        pred_X = np.expand_dims([np.array(dist[i])], axis=1)
        true_arr[i] = r2[i]
        del r2[i]
        del dist[i]

        model = LinearRegression()
        model.fit(np.expand_dims(np.array(dist), axis=1), r2)
        pred_arr[i] = model.predict(pred_X)
    _,p_ = icn_permutation_test.permutationTestSpearmansRho( true_arr, pred_arr, True, r"$R^2$", 10000)
    
    return true_arr, pred_arr, p_
```

```python
ECOG_CON_true, ECOG_CON_pred, ECOG_CON_p = leave_one_ch_out(loc="ECOG", lat_="r2_con")
ECOG_IPS_true, ECOG_IPS_pred, ECOG_IPS_p = leave_one_ch_out(loc="ECOG", lat_="r2_ips")
STN_CON_true, STN_CON_pred, STN_CON_p = leave_one_ch_out(loc="STN", lat_="r2_con")
STN_IPS_true, STN_IPS_pred, STN_IPS_p = leave_one_ch_out(loc="STN", lat_="r2_ips")
```

```python
plt.figure(figsize=(6,6), dpi=300)
plt.subplot(221)
sns.regplot(x=ECOG_CON_true, y=ECOG_CON_pred)
plt.ylabel(r"$R^2$"+ " predict"); plt.xlabel(r"$R^2$"+ " true"); 
plt.title("ECOG CON p="+str(np.round(ECOG_CON_p,2)))
plt.subplot(222)
sns.regplot(x=ECOG_IPS_true, y=ECOG_IPS_pred)
plt.ylabel(r"$R^2$"+ " predict"); plt.xlabel(r"$R^2$"+ " true"); 
plt.title("ECOG IPS p="+str(np.round(ECOG_IPS_p,2)))
plt.subplot(223)
sns.regplot(x=STN_CON_true, y=STN_IPS_pred)
plt.ylabel(r"$R^2$"+ " predict"); plt.xlabel(r"$R^2$"+ " true"); 
plt.title("STN CON p="+str(np.round(STN_CON_p,2)))
plt.subplot(224)
sns.regplot(x=STN_CON_true, y=STN_IPS_pred)
plt.ylabel(r"$R^2$"+ " predict"); plt.xlabel(r"$R^2$"+ " true"); 
plt.title("STN IPS p="+str(np.round(STN_IPS_p,2)))
plt.tight_layout()
```

### leave one patient out CV

```python
def leave_one_patient_out(loc="ECOG", lat_="r2_con"):
    pred_arr = []
    true_arr = []
    for sub in subjects:
        df_test = df_all[(df_all["sub"] == int(sub)) & (df_all["loc"] == loc)]
        df_train = df_all[(df_all["sub"] != int(sub)) & (df_all["loc"] == loc)]
        df_test[lat_]


        model = LinearRegression()
        model.fit(np.expand_dims(df_train["dist_con"], axis=1), df_train[lat_])
        pred_arr.append(model.predict(np.expand_dims(df_test["dist_con"], axis=1)))
        for i in range(df_test.shape[0]):
            true_arr.append(df_test[lat_].iloc[i])
    _,p_ = icn_permutation_test.permutationTestSpearmansRho(np.concatenate(pred_arr), \
                                                            np.array(true_arr), True, r"$R^2$", 10000)

    return np.array(true_arr), np.concatenate(pred_arr), p_
```

```python
ECOG_CON_true, ECOG_CON_pred, ECOG_CON_p = leave_one_patient_out(loc="ECOG", lat_="r2_con")
ECOG_IPS_true, ECOG_IPS_pred, ECOG_IPS_p = leave_one_patient_out(loc="ECOG", lat_="r2_ips")
STN_CON_true, STN_CON_pred, STN_CON_p = leave_one_patient_out(loc="STN", lat_="r2_con")
STN_IPS_true, STN_IPS_pred, STN_IPS_p = leave_one_patient_out(loc="STN", lat_="r2_ips")
```

```python
plt.figure(figsize=(6,6), dpi=300)
plt.subplot(221)
sns.regplot(x=ECOG_CON_true, y=ECOG_CON_pred)
plt.ylabel(r"$R^2$"+ " predict"); plt.xlabel(r"$R^2$"+ " true"); 
plt.title("ECOG CON p="+str(np.round(ECOG_CON_p,2)))
plt.subplot(222)
sns.regplot(x=ECOG_IPS_true, y=ECOG_IPS_pred)
plt.ylabel(r"$R^2$"+ " predict"); plt.xlabel(r"$R^2$"+ " true"); 
plt.title("ECOG IPS p="+str(np.round(ECOG_IPS_p,2)))
plt.subplot(223)
sns.regplot(x=STN_CON_true, y=STN_IPS_pred)
plt.ylabel(r"$R^2$"+ " predict"); plt.xlabel(r"$R^2$"+ " true"); 
plt.title("STN CON p="+str(np.round(STN_CON_p,2)))
plt.subplot(224)
sns.regplot(x=STN_CON_true, y=STN_IPS_pred)
plt.ylabel(r"$R^2$"+ " predict"); plt.xlabel(r"$R^2$"+ " true"); 
plt.title("STN IPS p="+str(np.round(STN_IPS_p,2)))
plt.tight_layout()
```
