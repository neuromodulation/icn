```python
class Generator():
    def __init__(self,data, label, window, stride):
        self.data = data
        self.label = label
        self.window = window
        self.stride = stride
        self.numSamples = self.data.shape[0]
        
        
    def generator(self, passes = np.inf):
        
        epochs = 0
        X_total = []
        y_total = []        
        while epochs < passes:

            for i in np.arange(0,self.numSamples-self.window,self.stride) :
                X_total.append(self.data[i: i+self.window])
                y_total.append(self.label[i: i+self.window])


            yield (X_total, y_total)
                
            epochs += 1
            print("epochs:",epochs)
                
        

        
```

#### window 1 ms, stride 1 sample

```python
window_ms = 5
window = int(window_ms* raw.info['sfreq'])
train_gen = Generator(X_train, y_train, window, 1000)
test_gen = Generator(X_test, y_test, window, 1000)


X_train_ep = np.array(next(train_gen.generator())[0])
y_train_ep = np.array(next(train_gen.generator())[1])
X_test_ep = np.array(next(test_gen.generator())[0])
y_test_ep = np.array(next(test_gen.generator())[1])

print(X_train_ep.shape)
print(y_train_ep.shape)

print(X_test_ep.shape)
print(y_test_ep.shape)
```

```python
def get_indxs(label):
    
    return np.where((np.diff(y>0)>0)==True)[1]

def get_epochs(data, label, length):
    
    
    indexes = get_indxs(label)
    class_0 = []
    class_1 = []
    
    label_0 = []
    label_1 = []
    
    for indx in indexes:
        class_0.append(data[indx:indx+length])
        class_1.append(data[indx-length:indx])
        
        label_0.append(label.T[indx:indx+length])
        label_1.append(label.T[indx-length:indx])
        
        
    class_0 = np.array(class_0)
    class_1 = np.array(class_1)
    
    label_0 = np.array(label_0)
    label_1 = np.array(label_1)

    
    
    return np.vstack([class_0,class_1]), np.vstack([label_0, label_1])
```
